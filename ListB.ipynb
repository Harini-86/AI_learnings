"""Beginners Level"""

```python
#Largest element
#Method1:
nums=[16,7,21,46,3,55]
largest = max(nums)
print(largest)

#Method2:
nums=[16,7,21,46,3,55]
largest=nums[0] #assume first is largest
for num in nums:
    if num>largest:
        largest = num
print(largest)

#Method3:
mylist=[20,13,85,4,33]
mylist.sort()
print(mylist[-1])
```

```python
#Smallest element
#Method1:
a=[23,45,2,18,50]
smallest = min(a)
print(smallest)

#Method2:
smallest=a[0] #assume first is largest
for num in a:
    if num<smallest:
        smallest = num
print(smallest)

#Method3:
a.sort()
print(a[0])
```

```python
#sum of all elements
#Method1:
nums = [2,3,14,29]
add = sum(nums)
print(add)

#Method2:
total = 0
for num in nums:
    total += num
print(total)

#Method3:
total = 0
i = 0
while i<len(nums):
    total += nums[i]
    i += 1
print(total)
```

```python
#Average of an elements
#Method1:
a=[3,5,8,12]
average = sum(a)/len(a)
print(average)

#Method2:
total = 0
for num in a:
    total += num
average = total/len(a)
print(average)

#Method3:
import statistics
b = [4,23,6,8]
average = statistics.mean(b)
print(average)
```

```python
#even and odd
#Method1:
a = [1,4,15,22,50,43,65]
even_count = 0
odd_count = 0
for num in a:
    if num%2 == 0: #even if remainder is 0
        even_count += 1
    else:
        odd_count += 1 #odd if remainder is 1
print("Even numbers:", even_count)
print("Odd numbers:", odd_count)

#Method2:
even_count = len([n for n in a if n%2 == 0])
odd_count = len([n for n in a if n%2!= 0])
print(even_count, odd_count)
        
```

```python
#Max and Min values without using max(),min()
num = [14,56,2,89,45,23,90,1]
max_val = num[0]
min_val = num[0]
for number in num:
    if number > max_val:
        max_val=number
    if number < min_val:
        min_val=number
print("Maximum:", max_val)
print("Minimum:", min_val)
```

```python
#reverse a list
#Method1:
nums = [4,6,20,3]
nums.reverse()
print(nums)

#Method2:
result = nums.reverse()
print(result)
print(nums)

#without using reverse() or slicing([::-1])
#Method1:
nums = [2,3,4,5,6]
left,right = 0,len(nums)-1

while left<right:
    nums[left],nums[right] = nums[right],nums[left] #swap
    left += 1
    right -= 1
print(nums)

#Method2:
nums = [4,6,2,7,10]
reversed_nums = []
for i in range(len(nums)-1,-1,-1):
    reversed_nums.append(nums[i])
print(reversed_nums)
```

```python
#Positive and Negative numbers
#Method1:
list = [12,-9,-45,57,2]
for num in list:
    if num>0:
        print(num,"is positive")
    else:
        print(num,"is negative")

#Method2:
list = [12,-9,-45,57,0,2]
positive = []
negative = []
zeros = [] #0 neither positive nor negative
for num in list:
    if num>0:
        positive.append(num)
    elif num<0:
        negative.append(num)
    else:
        zeros.append(num)
print("Positive numbers:",positive)
print("Negative numbers:",negative)
```

```python
#Counting zeros in a list
#Method1:
mylist = [4,0,6,0,0,56,34]
count = 0
for num in mylist:
    if num == 0:
        count += 1
print(count)

#Method2:
print("Number of zeros:", mylist.count(0))
```

```python
#Product of all elements
#Method1:
List = [3,5,11,7,8]
product = 1
for num in List:
    product *= num
print(product)  

#Method2:
import math
nums = [2,3,6,8]
print(math.prod(nums))
```

"""Intermediate Level"""

```python
#second largest element
#Method1:
nums = [12, 45, 2, 8, 4, 33]
nums = list(set(nums))
nums.sort()
print("Second largest:", nums[4])

#Method2:
nums = [34,65,3,45,80]
largest = max(nums)
nums.remove(largest) # remove only one largest
second_largest = max(nums)
print("Second largest:",second_largest)
```

```python
#second smallest element
#Method1:
L=[2,45,3,65,3]
L=list(set(L))
L.sort()
print("Second smallest:",L[1])

#Method2:
smallest = min(L)
L.remove(smallest)
second_smallest = min(L)
print("Second smallest:", second_smallest)
```

```python
#Removing duplicate elements while preserving original order
#Method1:
L=[2,45,3,65,3]
L=list(set(L))
print(L)

#Method2:
nums = [12,34,14,45,34,54,34,9,5,12]
unique = []
for num in nums:
    if num not in unique:
        unique.append(num)
print(unique)

#Method3:
unique = list(dict.fromkeys(nums))
print(unique)
```

```python
#All duplicate elements
nums = [22,23,45,12,8,23,45,7]
seen = set()
duplicates = []
for num in nums:
    if num in seen and num not in duplicates:
        duplicates.append(num)
    else:
        seen.add(num)
print(duplicates)

#Method2:
from collections import Counter

nums = [12, 45, 2, 8, 45, 33, 12, 2]
duplicates = [num for num, count in Counter(nums).items() if count > 1]
print(duplicates)
    
```

```python
#sorted in ascending order
#Method1:
numbers = [2, 4, 6, 8, 10]
is_sorted = True

for i in range(len(numbers) - 1):
    if numbers[i] > numbers[i + 1]:
        is_sorted = False
        break

if is_sorted:
    print("The list is sorted in ascending order.")
else:
    print("The list is not sorted in ascending order.")

#Method2:
numbers = [1, 2, 3, 6, 5]
sorted = True

for i in range(len(numbers) - 1):
    if numbers[i] > numbers[i + 1]:
        sorted = False

if sorted:
    print("List is sorted")
else:
    print("List is not sorted")
```

```python
#Merging two list without using extend()
#Method1:
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2
print(result)

#Method2:
list1 = [1, 28, 3]
list2 = [4, 7, 62]
for i in list2:
    list1.append(i)
print(list1)
```

```python
#Rotating the list to the left by one position
nums = [1, 2, 3, 4, 5]
nums = nums[1:] + nums[:1] 
print(nums) 

#Method2:
numbers = [6,87,45,13,16,9]
first = numbers[0]
for i in range(len(numbers) - 1):
    numbers[i] = numbers[i + 1]

numbers[-1] = first
print(numbers)
```
