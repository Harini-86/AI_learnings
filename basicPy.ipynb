#Operators

```python
#Arithmetic

a=20
c=56
print(a+c) #Addition
print(a-c) #Subtraction
print(a*c) #Multiplication
print(a/c) #Division
print(a//c) #Floor Division
print(a%c) #Modulus/Remainder
print(a**c) #Exponent
```

```python
#Comparison

a=45
b=48
print(a>b) #Greater than
print(a<b) #Less than
print(a==b) #Equal to
print(a!=b) #Not equalto
print(a>=40) #Greater than or equalto
```

```python
#Logical 

x=True
y=False
print(x and y) #Logical AND
print(x or y) # logical OR
print(not y) #Logical NOT
```

```python
#Assignment

p = int(input())
q = int(input())
p+=1
print(p)
q-=2
print(q)
p/=q
print(p)


```

```python
#membership

s = (54,"g",9,4.0,{2,5,7})
print(54 in s)
print(5 in s)
print(6 not in s)
print(5 in s[4])
a = {1,2,3}
print(1 in a)
```
