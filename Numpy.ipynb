```python
#numpy
import numpy as np
```

```python
#creating a numpy array from a list
arr1 = np.array([2,5.9,3,7])
print(arr1) #list into numpy array
```

```python
#creating a 2D Numpy array
arr2 = np.array([[6,5,4,3],[7,2,1,0]])
print(arr2)
```

```python
#Finding shape of an array
print(arr2.shape) #no. of rows and colums 
```

```python
##Finding data type of an element
print(arr1.dtype) #shows the datatype 
```

```python
#create an array of zeros
zeros=np.zeros((4,3))
print(zeros) 
```

```python
#create an array of ones
ones = np.ones((2,3))
print(ones)
```

```python
#create array with a range of numbers
N = np.arange(2,80,10)
print(N) #creates an array from 2 to 80 with a step of 10
```

```python
#create an array with evenly spaced values
E = np.linspace(1,10,5)
print(E) #creates 5 equally spaced numbers b/w 1 to 10
```

```python
#re-shape an  array
print(N.reshape(4,2)) 
#reshapes a 1D array of 8 elements into a 4x2 matrix
```

```python
arr=np.arange(4).reshape(2,2)
print(arr)
```

```python
#flatten a multidimensional array
print(arr2.flatten()) #converts a muldim into 1D array
```

```python
# Find max min and sum of all elements
print(arr2.max())
print(arr.min())
print(arr1.sum())
```

```python
#Finding mean,median,sd
print(arr1.mean())
print(np.median(arr)) #There is no median attribute in numpy
print(arr2.std())
```

```python
#slice an array
print(arr1[3:])
print(arr2[::2])
```

```python
#Accessing element element in a 2D array
print(arr2[0,2])
print(arr2[1,0]) #shows 2nd row,3rd column
```

```python
#perform element-wise addition
arr3 = np.array([4,5,6,10])
arr4 = np.array([45,3,2,0])
print(arr3 + arr4) #adds  corresponding elements
```

```python
#perform element-wise multiplication
print(arr3 * arr4)
```

```python
#Matrix multiplication
print(np.dot(arr1,arr1))

matx1 = np.array([5,4,5])
matx2 = np.array([6,10,0])
print(matx1*matx2)
print(np.dot(matx1,matx2))
```

```python
#Transpose of a matrix
print(arr2.T)
```

```python
#Finding unique elements in array
w = np.array([4,5,2,4,1,7,2,9,4])
print(np.unique(w))
print(np.unique(matx1))
```

```python
#using boolean indexing
print(w[w>5])
print(arr[arr<3])
```

```python
#replace elements based on condition
w[w<4] = 0
print(w)

h=np.array([3,2,5,76,67,87,90])
h[h<76]=1
print(h)
```

```python
#Generate random integers
INTEGER = np.random.randint(19,48,10)
print(INTEGER)
```

```python
#Generate random floats
FLOAT = np.random.rand(2,1)
print(FLOAT)
```

```python
#Shuffle elements of an array
np.random.shuffle(arr2)
print(arr2)
```

```python
#Sort an array
V=np.sort(INTEGER)
print(V)
```

```python
#Stack array vertically and horizontally
print(np.vstack((matx1,matx2)))
print(np.hstack((matx1,matx2)))
```

```python
ver = np.array(([4,5,6],[2,8,1]))
hor = np.array(([6,7,4],[1,4,6]))
print(np.vstack(ver))
print(np.hstack((ver,hor)))
```

```python
#finding correlation coefficient
print(np.corrcoef((matx2,matx1)))
```
